```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

service = Service("/usr/local/bin/chromedriver")
options = webdriver.ChromeOptions()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=service, options=options)

place_url = "https://2gis.kz/almaty/firm/9429940000799837/tab/reviews"
driver.get(place_url)
time.sleep(5)

num_reviews_before = 0
scroll_pause_time = 3

while True:
    review_containers = driver.find_elements(By.CLASS_NAME, "_1k5soqfl")
    num_reviews = len(review_containers)

    if num_reviews == num_reviews_before:
        break

    num_reviews_before = num_reviews
    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
    time.sleep(scroll_pause_time)

print(f"Total reviews found: {len(review_containers)}")

star_counts = []

for index, container in enumerate(review_containers):
    try:
        star_container = container.find_element(By.CLASS_NAME, "_1fkin5c")

        if star_container:
            print(f"Found star container for Review {index+1}")

        star_elements = star_container.find_elements(By.TAG_NAME, "span")
        print(f" Review {index+1} has {len(star_elements)} stars")

        star_counts.append(len(star_elements))
    except:
        print(f"No star container found for Review {index+1}")
        star_counts.append(0)

driver.quit()

for i, stars in enumerate(star_counts):
    print(f"Review {i+1}: {stars} stars")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# Set up ChromeDriver
service = Service("/usr/local/bin/chromedriver")
options = webdriver.ChromeOptions()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=service, options=options)

# Target page URL
place_url = "https://2gis.kz/almaty/firm/70000001017515451/76.914432%2C43.219634/tab/reviews?m=76.909516%2C43.233886%2F14"
driver.get(place_url)
time.sleep(5)  # Allow page to load

# Scroll down to load all reviews
num_reviews_before = 0
scroll_pause_time = 3

while True:
    reviews = driver.find_elements(By.CLASS_NAME, "_1k5soqfl")
    num_reviews = len(reviews)

    if num_reviews == num_reviews_before:
        break  # Stop if no new reviews are loaded

    num_reviews_before = num_reviews
    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
    time.sleep(scroll_pause_time)

# Extract reviews and star ratings
reviews_list = []
star_counts = []

review_elements = driver.find_elements(By.CLASS_NAME, "_1k5soqfl")

for element in review_elements:
    try:
        # Extract review text
        review_text_elements = element.find_elements(By.CLASS_NAME, "_1msln3t")
        review_text = " ".join([e.text for e in review_text_elements]).strip()

        # Extract star rating
        try:
            star_container = element.find_element(By.CLASS_NAME, "_1fkin5c")
            star_elements = star_container.find_elements(By.TAG_NAME, "span")
            star_count = len(star_elements)  # Count stars
        except:
            star_count = 0  # If no stars found, assume 0 stars

        # Save extracted data
        reviews_list.append(review_text)
        star_counts.append(star_count)

    except Exception as e:
        print(f"Error processing review: {e}")

driver.quit()

# Save data to DataFrame
df = pd.DataFrame({"Review": reviews_list, "Stars": star_counts})

# Display extracted data
print(df.head())

# Optionally save to CSV
df.to_csv("reviews_data.csv", index=False)
```

```python
def assign_label(stars):
    if stars > 3:
        return "positive"
    elif stars == 3:
        return "neutral"
    else:
        return "negative"

df["label"] = df["Stars"].apply(assign_label)
```
